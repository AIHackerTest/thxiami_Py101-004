### 0.2 前后端交互——HTML和Jinja2

搭建起一个服务器, 它需要有前端和后端。

- 前端是什么？

  我们这次的天气查询系统，用户打开浏览器，输入地址后，看到了网页上的内容，并且可以在网页上输入城市进行查询，或者点击按钮。与用户直接交互的这部分就是前端，其功能是与用户交互，然后把用户的请求发送给后端，获得后端返回的结果并展示给用户。


- 后端是什么？

    比如，我们使用Flask运行起的程序，在从前端获得用户输入的城市或点击的按钮后，如果是查询天气，那就调用API查询，然后把需要的数据传给前端。后端的功能是处理前端发来的请求，然后把处理的结果返回给前端。

所以在之前的ch1和ch2任务中，我们完成了在命令行运行的天气查询程序，通过在命令行输入城市或指令发起请求，结果通过`print()`直接显示在命令行。这次我们做的就是，让用户通过网页发起请求，通过网页看到结果。
通过复用ch1和ch2的任务成果, 我们已经完成后端天气查询的核心代码，所以为了完成任务，应该学习下前后端如何交互，前端如何显示结果。

回顾下我们之前的例子，当程序运行后，用户打开浏览器,输入网址确认后,浏览器显示页面，我们可以实现显示"Hello world"这样简单的文字, 但如果需要显示更复杂的内容,该怎么办?这里就要引入**HTML**和**Jinja2**

- HTML

    前端网页的显示需要用到浏览器+HTML文件，HTML文件它是一个纯文本文件，按一定的格式组成，其功能是让浏览器在读取并显示它时，按照格式对应的方式显示。就好像markdown一样。
    关于HTML可参见:
    - [Introduction to HTML](https://www.w3schools.com/html/html_intro.asp)
    - [W3school 中文站](http://www.w3school.com.cn/html5/index.asp)


- Jinja2

    假如编辑好了页面, 如何能返回给前端,让浏览器能显示，之前的Flask例子中,用户访问网页时,我们返回了"Hello world",那如果想让前端显示我们编辑好的 HTML 页面, 该怎么办呢?
    使用Flask内置的 jinja2 模板引擎就可以轻松实现。参考[rendering-templates](http://flask.pocoo.org/docs/0.12/quickstart/#rendering-templates)可快速学习。

    以下是示范代码：

````python
from flask import (
    Flask,
    render_template,
)


app = Flask(__name__)

@app.route('/')
def hello_world():
    return render_template('index.html')

app.run(
    host='0.0.0.0',
    port=3000
)
````

范例讲解：

render_template('index.html') 内的index.html就是我们编辑好的HTML文件，使用这个函数可以把页面返回给前端。

使用时需注意index.html必须放在指定位置才能生效

假如我们上面的代码文件叫main.py, 其正确位置见下方。

```
app/
    main.py
    templates/
        index.html
```

### 0.3 前后端交互——前端如何发出请求

怎么样才能让用户在网页上进行查询天气，或者查询历史记录/帮助。这也内含了我们怎么知道用户是查询天气还是查询其他内容。

经搜索，已知的办法有：
- 1，利用HTML的form
- 2，利用JS中AJAX技术

第一种技术比较简单，先使用第一种技术实现。以下分别为前端的HTML代码和后端的代码。

- 前端，index.html

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>问天内测版</title>
</head>
<body>
<form align="center" class="form-inline" action="/search" method="GET">
    <input type="text" class="form-control" name="user_input" placeholder="请输入城市名(中文或拼音)">
    <button type="submit">查询天气</button>
</form>

<form align="center" action="/command" method="GET">
    <button type="submit" name="button" value="history">历史记录</button>
    <button type="submit" name="button" value="help">帮助</button>
</form>
</body>

- 后端

````python
from flask import (
    Flask,
    request,
    render_template,
)


@app.route('/')
def hello_world():
    return render_template('index.html')


@app.route('/search', methods=['GET'])
def search():
    """
    天气查询的路由函数
    """
    query = request.args
    user_input = query.get('user_input', '')
    print('用户输入了:', user_input)
    

@app.route('/command', methods=['GET'])
def command_response():
    """
    处理用户点击 帮助/历史记录 button 的路由函数
    """
    command = request.args.get('button', None)
    print('用户点击了{}按钮'.format(command))
    
    
app.run(
    debug=True,  # 这里为True时, 当修改了后端代码,不需要停止程序然后重新运行,修改就能生效
    host='0.0.0.0',
    port=3000
)
````

### 0.4 前后端交互——后端如何传数据给前端

至此, 我们在后端就能知道用户点击了什么按钮，或者输入了什么内容，但上面的例子还不完整，缺少了处理请求，并将数据返回给前端的环节。
处理请求这一步，我们复用ch2的成果稍作修改即可。关键是如何将结果传给前端并显示。
这里我们同样利用Jinja2模板可以实现。下面是一个伪代码例子。

**后端**

````python
@app.route('/search', methods=['GET'])
def search():
    """
    天气查询的路由函数
    """
    query = request.args
    user_input = query.get('user_input', '')
    # 假设get_weather()函数传入查询的城市,返回天气查询结果,这里只考虑查询正常的情况
    weather_dict = get_weather(location=user_input)  
    return render_template('results.html', results=weather_dict)
````

通过 render_template('results.html', results=weather_dict) ,把变量"传给"results.html。

编辑 results.html 时，使用`<p>城市: {{ results.city }} </p>` 这种形式， 当 jinja2 返回页面给前端时，它会自动使用 `results[dict]` 的值来代替 `{{ results.city }}`,这样使得前后端交互变得简单，也使得二者更加隔离：后端返回数据给前端，前端拿数据生成HTML页面。

此外，Jinja2 还支持 for语句,这可用于显示历史记录。更多用法可参见:[Welcome to Jinja2](http://jinja.pocoo.org/docs/dev/)

                1. **Expires**：从图可以看出，expires的值是一个绝对时间，是http1.0的功能。如果浏览器的时间没有超过这个expires的时间，代表缓存还有效，命中强缓存，直接从缓存读取资源。不过由于存在浏览器和服务端时间可能出现较大误差，所以在之后http1.1提出了cache-control。

                2. **Cache-Control**：从图可以看出，Cache-Control的值是类似于max-age=43200这样的，是一个相对时间，43200是秒数，正好是12小时。当浏览器第一次请求资源的时候，会把response header的内容缓存下来。之后的请求会先从缓存检查该response header，通过第一次请求的date和cache-control计算出缓存有效时间。如果浏览器的时间没有超过这个缓存有效的时间，代表缓存还有效，命中强缓存，直接从缓存读取资源。

                3. 两者可以同时设置，但是HTTP1.1中判断优先级: cache-control > expires

        根据[你应该知道的浏览器缓存知识](https://excaliburhan.com/post/things-you-should-know-about-browser-cache.html)这篇博客的内容:
        > 那么如果资源有更改的时候，怎么让浏览器知道呢？现在常用的解决方法是加一个?v=xxx的后缀，在更新静态资源版本的时候，更新这个v的值，这样相当于向服务端发起一个新的请求，从而达到更新静态资源的目的。

        与此对应, 本次程序中这么改:
        ```html
        <script src="{{ url_for('static', filename='options.js')?v=1709091730 }}"></script>
        ```
        上面这种方法是普遍做法, 它有2个好处:
            - 保存了缓存机制以优化性能
            - 当静态文件发生变化时客户端又能重新获取最新资源,不会获取过期资源
            
        - 缩短服务端的缓存有效期
            1. 按照Stack Overflow上的办法修改`app.config['SEND_FILE_MAX_AGE_DEFAULT'] = 300`后, 重新打开浏览器或强行刷新(cmd+shift+R)后,`option.js`的响应头如下,我们发现`max-age=300`:
            ![image](https://user-images.githubusercontent.com/26535231/30242071-383fa89e-95c2-11e7-9a1c-de9bc2a7c4f1.png)
            
            说明修改办法有效, 但不太推荐使用该方法将缓存有效时间修改为很短, 这样不利于上线后的性能优化。在这里探索只是想搞明白Flask服务端该如何修改缓存有效时间。
        - 禁用浏览器缓存机制
        通过增加以下代码,可以在处理收到的所有request后返回的响应头中加入`'Cache-Control' = 'no-store'`,其作用是让浏览器不缓存请求或响应的任何内容。
        ```python
        @app.after_request
        def add_header(response):
            # response.cache_control.no_store = True
            response.headers['Cache-Control'] = 'no-store'
            return response
        ```
        增加后测试，我们发现`option.js`的`response headers: Cache-Control: no-store`,重新访问页面时可以看到`option.js`也重新加载,不再是`from memory cache`
        
        上面那段代码中注释掉了一行,其也可以修改返回的response.headers中的cache_control,如果使其生效并注释掉原来那行代码
        ```python
        @app.after_request
        def add_header(response):
            response.cache_control.no_store = True
            # response.headers['Cache-Control'] = 'no-store'
            return response
        ```
        
        这时拿到的`option.js`的`response headers: Cache-Control:public, max-age=43200, no-store,增加了一项no-store,但原来的max-age=43200并没有去掉,不过经测试此时`no-store`已经生效,重新访问页面时可以看到`option.js`也重新加载
